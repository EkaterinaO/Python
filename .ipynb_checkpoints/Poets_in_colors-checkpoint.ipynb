The main idea is to plot all colors, which poets use in their poems. 

```python
import gutenberg
import numpy as np
```

```python
from gutenberg.acquire import load_etext
from gutenberg.cleanup import strip_headers
```

Getting poems from gutenberg library, by their numbers. 

```python
from gutenberg.acquire import load_etext
from gutenberg.cleanup import strip_headers
#Poems, by Alexander Pushkin
text_Pushkin = strip_headers(load_etext(54991)).strip()
#Poems, by Victor Hugo
text_Hugo = strip_headers(load_etext(8775)).strip()

text_Eliot = strip_headers(load_etext(1567)).strip()

text_Sterling = strip_headers(load_etext(56764)).strip()


text_Blake = strip_headers(load_etext(574)).strip()


text_Browning = strip_headers(load_etext(16376)).strip()


text_Dickinson = strip_headers(load_etext(2678)).strip()
```

```python
# getting all colors which we can visualize with python matplotlib
from __future__ import division
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors


colors = dict(mcolors.BASE_COLORS, **mcolors.CSS4_COLORS)

by_hsv = sorted((tuple(mcolors.rgb_to_hsv(mcolors.to_rgba(color)[:3])), name)
                for name, color in colors.items())
sorted_names = [name for hsv, name in by_hsv]
```

Caounting the words and assosiating them with their colors. 

```python
def word_count(string):
    my_string = string.lower().split()
    my_dict = {}
    colors =  []
    for item in my_string:
        item.strip('\r,\n,\t,.,-,--')
        for color in sorted_names:
            if (item == color):
                #print(color)
                colors.append(color)
    return colors


Colors = [ word_count(text_Pushkin), word_count(text_Hugo),word_count(text_Eliot), word_count(text_Sterling), word_count(text_Blake),
word_count(text_Browning),
word_count(text_Dickinson)]
Authors = ['ALEXANDER PUSHKIN', 'Victor Hugo','T. S. ELIOT', 'GEORGE STERLING', 'William Blake', 'Robert Browning', 'EMILY DICKINSON']
```

Ploting all the colors, which author used. 

```python
for _ in range(len(Colors)):
    rng = np.random.RandomState(0)
    x = rng.randn(len(Colors[_]))
    y = rng.randn(len(Colors[_]))

    sizes = 700 
    
    plt.scatter(x, y, c = Colors[_], s = sizes,  alpha=0.7,
                   cmap='viridis')
    plt.title(Authors[_])
    plt.show()
```
